## Import Dependencies

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from statistics import mode
```

## Load the Data

```python
df = pd.read_csv("Data/train.csv", header=None)
test_data = pd.read_csv("Data/test.csv", header=None)
train_labels = pd.read_csv("Data/trainLabels.csv", names = ["Label"])
```

```python
print("Shape is: ", df.shape)
df.head()
```

```python
print("Shape is: ", test_data.shape)
test_data.head()
```

```python
print("Shape is: ", train_labels.shape)
train_labels.head()
```

## Prepare Data

```python
df = pd.concat([df, train_labels], axis=1)
print("Shape is: ", df.shape)
df.head()
```

## Check Missing Values

```python
df.isnull().sum()
```

```python
test_data.isnull().sum()
```

## Train-Test Split

```python
predict = "Label"
df_train = df.drop(columns=[predict])
y = df[[predict]].values.ravel()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_train, y, test_size=0.20, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

## Build the Model

```python
from sklearn import svm
model_1 = svm.SVC()
model_1.fit(X_train, y_train) 
print(model_1.score(X_test, y_test))
```

```python
from catboost import CatBoostClassifier
model_2 = CatBoostClassifier()
model_2.fit(X_train, y_train) 
print(model_2.score(X_test, y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_3 = GradientBoostingClassifier()
model_3.fit(X_train, y_train) 
print(model_3.score(X_test, y_test))
```

```python
from lightgbm import LGBMClassifier
model_4 = LGBMClassifier()
model_4.fit(X_train, y_train) 
print(model_4.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
model_5 = RandomForestClassifier()
model_5.fit(X_train, y_train) 
print(model_5.score(X_test, y_test))
```

## Test The Model

### Final Model is an Ensemble of all 4 ML models with Mode of Output of all Models taken, i.e. Majority Voting

```python
test_data
```

```python
X_train
```

```python
results_df = pd.DataFrame(columns = ["Prediction"])
count = 0
for index, row in test_data.iterrows():
    predictions = []
    count = count + 1
    row = row.to_frame()
    row = row.T
    prediction_1 = model_1.predict(row)
    prediction_2 = model_2.predict(row)
    prediction_3 = model_3.predict(row)
    prediction_4 = model_4.predict(row)
    prediction_5 = model_5.predict(row)
    predictions.append(prediction_1.tolist()[0])
    predictions.append(prediction_2.tolist()[0])
    predictions.append(prediction_3.tolist()[0])
    predictions.append(prediction_4.tolist()[0])
    predictions.append(prediction_5.tolist()[0])
    print(count, predictions)
    to_append = int(mode(predictions))
    a_series = pd.Series(to_append, index = results_df.columns)
    results_df = results_df.append(a_series, ignore_index=True)
```

```python
results_df.columns = ['Solution']
results_df['Id'] = np.arange(1,results_df.shape[0]+1)
results_df = results_df[['Id', 'Solution']]
```
